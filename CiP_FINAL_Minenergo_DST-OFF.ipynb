# Описание решения
## Вводная информация
1. На сайте Системного оператора Единой энергетической систем (ЕЭС) России представлена информация о почасовом потреблении электроэнергии в рамках объединенных энергетических систем (ОЭС) https://br.so-ups.ru/BR/GenConsum. Необходимо обеспечить автоматическое скачивание этой информации.
2. Надо выбрать социально-экономические и другие показатели по территориям и отраслям и проверить их на корреляцию с уровнем потребления электроэнергии. Эти показатели должны иметь регулярно обновляемые датасеты. Нужно обеспечить их автоматическую загрузку за период, сопоставимый с данными о потреблении электроэнергии.
3. Имея загруженные датасеты потребления электроэнергии и выбранных показателей, надо визуализировать зависимость между ними и смоделировать сценарии развития ситуации в будущем.

Примечания:
1. Важно учесть влияние сезонного и температурного факторов на потребление электроэнергии. Для этого необходимо загрузить информацию о дневной температуре по регионам.
2. Важно учесть влияние инфляции на показатели, имеющие денежное выражение. Для этого необходимо загрузить данные о ежемесячной инфляции.


## Структура хранения информации о потреблении электроэнергии
Для решения задачи вся информация должна быть структурирована с возможностью последующего масштабирования и с учетом использования базы данных.

**df_energy_db**:\
`index [int]` - номер записи\
`date_time [datetime]` - день и час регистрации потребления\
`period [str]` - периодичность регистрации показателя ('час')\
`sys_id_level [int]` - уровень учета - {0:'ЕЭС', 1:'ОЭС', 2:'Регион'}\
`power_sys_id [int]` - номер ОЭС или региона\
`unit_name  [int]` - название учетной единицы (региона, ОЭС)\
`e_use_fact [float]` - объем потребления электроэнргии\
`dt_input [datetime]` - дата и время загрузки информации\
`source [str]` - источник загрузки

## Структура хранения социально-экономических и иных показателей 
Необходимо структурировать информацию о социально-экономических показателях. Формат должен позволять легко добавлять новые показатели. В перспективе возможно использование базы данных.

**df_indicators_db**:\
`index [int]` - номер записи\
`indicator_name [str]` - название индикатора\
`unit_level [int]` - уровень учета - {0:'Россия', 1:'ОЭС', 2:'Регион'}\
`unit_name  [int]` - название учетной единицы (региона, ОЭС)\
`power_sys_id [int]` - номер ОЭС или региона\
`date_time [datetime]` - точка времени в интервале, в котором рассчитан показатель\
`period [str]` - периодичность регистрации показателя ('час', 'месяц', 'год')\
`value [float]` - результат активности в виде набора шкал с ответами\
`moment [datetime]` - дата и время загрузки индикатора\
`source [str]` - источник данных информации о значении индикатора

_Комментарий: `power_sys_id [int]` рассчитывается отдельно в привязке к ОЭС. Важно, что могут быть загружены данные даже по одному показателю из разных источников за разные периоды._

## Алгоритм решения

1. Загружаем исходные данные о потреблении электроэнергии, извлекаем и структурируем  информацию в приведенном выше формате **df_energy_db**
2. Загружаем исходные данные со значениями социально-экономических и других показателей из перечня доступных источников, извлекаем и структурируем их в приведенном выше формате **df_indicators_db**
3. Проверим корреляции показателей с данными о потреблении электроэнергии
4. В рамках модели прогнозирования сделаем среднесрочный прогноз. Сравним данные с информацией Минэкономразвития. Выделим обучающую и тестовую выборки, обучим модель и оценим качество прогноза.

```python
import os, wget

import numpy as np
import pandas as pd
import datetime as dt

pd.options.display.max_colwidth = 300

import matplotlib.pyplot as plt
import seaborn as sns

import re
import fitz
from tqdm import tqdm

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor 
from sklearn import metrics

RANDOM_SEED = 42

FILE_DATASET_E_USE_FACT = "Генерация и потребление_01.07.2010.csv"

# Словарь соответствий кодод ОЭС их названиям
oes_dict = {
    1019: 'ЕЭС России',
    530000: 'ОЭС Центра',
    540000: 'ОЭС Востока',
    550000: 'ОЭС Юга',
    600000: 'ОЭС Средней Волги',
    610000: 'ОЭС Сибири',
    630000: 'ОЭС Урала',
    840000: 'ОЭС Северо-Запада'
}

# Словарь принадлежности регионов ОЭС
dict_regions = {
    'Российская Федерация': 1019,
    'Центральный федеральный округ': 0,
    'Белгородская область': 530000,
    'Брянская область': 530000,
    'Владимирская область': 530000,
    'Воронежская область': 530000,
    'Ивановская область': 530000,
    'Калужская область': 530000,
    'Костромская область': 530000,
    'Курская область': 530000,
    'Липецкая область': 530000,
    'Московская область': 530000,
    'Орловская область': 530000,
    'Рязанская область': 530000,
    'Смоленская область': 530000,
    'Тамбовская область': 530000,
    'Тверская область': 530000,
    'Тульская область': 530000,
    'Ярославская область': 530000,
    'г.Москва': 530000,
    'Северо-Западный федеральный округ': 0,
    'Республика Карелия': 840000,
    'Республика Коми': 840000,
    'Архангельская область': 840000,
    'Вологодская область': 530000,
    'Калининградская область': 840000,
    'Ленинградская область': 840000,
    'Мурманская область': 840000,
    'Новгородская область': 840000,
    'Псковская область': 840000,
    'г.Санкт-Петербург': 840000,
    'Южный федеральный округ': 0,
    'Республика Адыгея': 550000,
    'Республика Калмыкия': 550000,
    'Республика Крым': 550000,
    'Краснодарский край': 550000,
    'Астраханская область': 550000,
    'Волгоградская область': 550000,
    'Ростовская область': 550000,
    'г. Севастополь': 550000,
    'Северо-Кавказский федеральный округ': 0,
    'Республика Дагестан': 550000,
    'Республика Ингушетия': 550000,
    'Кабардино-Балкарская Республика': 550000,
    'Карачаево-Черкесская Республика': 550000,
    'Республика Северная  Осетия - Алания': 550000,
    'Чеченская Республика': 550000,
    'Ставропольский край': 550000,
    'Приволжский федеральный округ': 0,
    'Республика Башкортостан': 630000,
    'Республика Марий Эл': 600000,
    'Республика Мордовия': 600000,
    'Республика Татарстан': 600000,
    'Удмуртская Республика': 630000,
    'Чувашская Республика': 600000,
    'Пермский край': 630000,
    'Кировская область': 630000,
    'Нижегородская область': 600000,
    'Оренбургская область': 630000,
    'Пензенская область': 600000,
    'Самарская область': 600000,
    'Саратовская область': 600000,
    'Ульяновская область': 600000,
    'Уральский федеральный округ': 0,
    'Курганская область': 630000,
    'Свердловская область': 630000,
    'Тюменская область': 630000,
    'Челябинская область': 630000,
    'Сибирский федеральный округ': 0,
    'Республика Алтай': 610000,
    'Республика Бурятия': 610000,
    'Республика Тыва': 610000,
    'Республика Хакасия': 610000,
    'Алтайский край': 610000,
    'Забайкальский край': 610000,
    'Красноярский край': 610000,
    'Иркутская область': 610000,
    'Кемеровская область': 610000,
    'Новосибирская область': 610000,
    'Омская область': 610000,
    'Томская область': 610000
}

# Функция для анализа содержимого датафреймов
def view_types(data, dropna=True):
    """
    Вывод отчета с анализом содержимого объекта data, подсчет представленных типов данных
    Parameters
    ----------
    data : DataFrame, Series, dict, list, другой тип, преобразуемый к DataFrame
        Объект DataFrame или приводимый к DataFrame для анализа
    dropna : bool, default True
        Не включать NaN в расчет количества уникальных значений
    Returns
    -------
    None
    2020 (c) Nikolay Ganibaev
    """
    from pandas import DataFrame
    data = DataFrame(data)
    columns_exch = {
        "<class 'str'>": 'str', "<class 'int'>": 'int', "<class 'float'>": 'float',
        "<class 'list'>": 'list', "<class 'dict'>": 'dict',
        "<class 'datetime.datetime'>": 'datetime',
        "<class 'pandas._libs.tslibs.timestamps.Timestamp'>": 'Timestamp',
    }
    df_output = pd.DataFrame(
        data[data[c].notna()][c].apply(type).value_counts() for c in data.columns
    ).fillna(0).astype(int)
    if data.isna().any().any():
        df_output['NaN'] = data.isna().sum()
    df_output['sum'] = df_output.sum(axis=1)
    df_output['(unique)'] = [data[c].astype(str).nunique(dropna) for c in data.columns]
    df_output.columns = [columns_exch.get(str(x), x) for x in df_output.columns]
    display(df_output.head(60))
    
    
# Функция расчета даты на начало следующего месяца для datetime
def next_month(d):
    return d.replace(d.year+1,1,1) if d.month == 12 else d.replace(month=d.month+1, day=1)

def load_indicator(INDICATOR, DIR_NAME, date_start=None):
    """
    Универсальная функция инкрементной загрузки значений индикаторов через API с сайта https://so-ups.ru/
    Parameters
    ----------
    INDICATOR: tuple
        (название индикатора, шаблон API-запроса, номер колонки со значением индикатора в csv-файле)
    DIR_NAME: str
        Код индикатора и имя директории для хранения csv-файлов
    date_start: str or datetime-like
        Дата периода загрузки индикатора, по умолчанию 2011-01-01
    Returns
    -------
    None
    2020 (c) Nikolay Ganibaev
    """
    if not os.path.isdir(DIR_NAME):
        os.mkdir(DIR_NAME)

    if date_start==None:
        date_start = dt.datetime(2011, 1, 1)

    # Загружаем по текущий момент
    date_end = next_month(dt.datetime.utcnow() + dt.timedelta(hours=+3))
    
    # Удаляем последние файлы в кэше
    cache_update = sorted(os.listdir(DIR_NAME))[-len(oes_dict):]
    
    for filename in cache_update:
        os.remove(f"{DIR_NAME}\\{filename}")

    for date_get in tqdm(pd.date_range(date_start, date_end, freq='M')):
        for ees_oes in oes_dict:
            csv_file = (f"https://so-ups.ru/index.php?id={DIR_NAME}&no_cache=1&tx_ms1cdu_pi1[kpo]={str(ees_oes)}&" +
                        f"tx_ms1cdu_pi1[dt]={date_get.strftime('%d.%m.%Y')}&tx_ms1cdu_pi1[format]=csv")
            
            if not f"{date_get.strftime('%Y-%m-%d')}_{INDICATOR[1]}_{str(ees_oes)}.csv" in os.listdir(DIR_NAME):
                wget.download(csv_file, DIR_NAME, None)
        date_get = (dt.datetime(date_get.year+1, 1, 1) 
                    if date_get.month == 12 
                    else date_get.replace(month=date_get.month+1, day=1))
        date_get = next_month(date_get)
```

# Прогнозная модель

## Прогноз потребления электроэнергии

```python
df
```

```python
# Разбиваем набор данных на тренировочный и тестовый
y = df['e_use_fact'].values
X = df[['mean_ipp','min_ipp','max_ipp','otrg_tov','stroit']]
X
```

```python
size_train = int(len(X)*0.1)

X_train = X.iloc[:-size_train]
X_test = X.iloc[size_train:]
y_train = y[:-size_train]
y_test = y[size_train:]

# Создаём модель, используя метод Random forest
model = RandomForestRegressor(n_estimators=100, verbose=1, n_jobs=-1, random_state=RANDOM_SEED)

# Обучаем модель на тестовом наборе данных
model.fit(X_train, y_train)

# Используем обученную модель для предсказания в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = model.predict(X_test)
```

```python
mae = metrics.mean_absolute_error(y_test,y_pred)
mean = df['e_use_fact'].mean()
mae_percent = mae/mean

print('среднее уровень потребления электроэнергии в прогнозируемом интервале:', mean)
print('средняя ошибка прогноза: ', mae)
print('средняя ошибка:', mae_percent*100, '%')
```
