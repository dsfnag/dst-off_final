# Постановка задачи
## Суть кейса
Понятие надежности электроснабжение - это основа функционирования как экономики, так и качества жизни в целом.

Возможность прогнозирования динамики потребления, возможность увязывания динамики потребления с экономическими показателями развития как регионов, так и отраслей, дает нам возможность более качественно планировать как ремонтные работы, так и другие программы, которые реализуются в рамках нашей отрасли.

Если мы сможем построить качественную, понятную, удобную, легкую модель, которая позволит нам не только прогнозировать, но  и моделировать различные сценарии развития как потребления электроэнергии, так и экономических показателей, мы получим окно возможностей для повышения качества работы как нашей отрасли, так и системы в целом.


## Что ожидается от участников
1. Не зацикливаться на чем-то одном. Возможно, не привязываться только к экономическим показателям. Возможно, есть другие вещи на которые стоит обратить внимание.
2. Где-то выйти за границы, где-то выйти в целом за поставленную задачу.
3. Надо подойти к задаче как с точки зрения кода, так и с точки зрения визуализации решения.

## Что будет считаться решением кейса
1. Перечень экономических и других показателей, коррелирующих с динамикой потребления электроэнергии по территориям и отраслям.
2. Определение взаимного влияния потребления электроэнергии и собранных показателей на основе ретроспективных данных.
3. Модель прогнозирования динамики показателей экономического роста России по отраслям и территориям, исходя из динамики потребления электроэнергии.
4. Платформенное решение на базе существующих систем, позволяющее проводить моделирование и прогнозирование онлайн.

## Сценарий использования результата

- Данные о потреблении электроэнергии и другие данные по API поступают в аналитический центр Минэнерго России, предложенная платформа в рамках разработанной модели обращается за необходимыми данными, проводит расчёт и прогнозирование экономических показателей, предоставляет его результаты в базу данных, после чего они представляются на дашборд и визуализируются для принятия управленческих решений.

Пример зависимости: динамика потребления электроэнергии и объем промышленного производства, динамика потребительской активности, изменения оборотов бизнеса.

# Описание решения
## Вводная информация
1. На сайте Системного оператора Единой энергетической систем (ЕЭС) России представлена информация о почасовом потреблении электроэнергии в рамках объединенных энергетических систем (ОЭС) https://br.so-ups.ru/BR/GenConsum. Необходимо обеспечить автоматическое скачивание этой информации.
2. Надо выбрать социально-экономические и другие показатели по территориям и отраслям и проверить их на корреляцию с уровнем потребления электроэнергии. Эти показатели должны иметь регулярно обновляемые датасеты. Нужно обеспечить их автоматическую загрузку за период, сопоставимый с данными о потреблении электроэнергии.
3. Имея загруженные датасеты потребления электроэнергии и выбранных показателей, надо визуализировать зависимость между ними и смоделировать сценарии развития ситуации в будущем.

Примечания:
1. Важно учесть влияние сезонного и температурного факторов на потребление электроэнергии. Для этого необходимо загрузить информацию о дневной температуре по регионам.
2. Важно учесть влияние инфляции на показатели, имеющие денежное выражение. Для этого необходимо загрузить данные о ежемесячной инфляции.


## Структура хранения информации о потреблении электроэнергии
Для решения задачи вся информация должна быть структурирована с возможностью последующего масштабирования и с учетом использования базы данных.

**df_energy_db**:\
`index [int]` - номер записи\
`date_time [datetime]` - день и час регистрации потребления\
`period [str]` - периодичность регистрации показателя ('час')\
`sys_id_level [int]` - уровень учета - {0:'ЕЭС', 1:'ОЭС', 2:'Регион'}\
`power_sys_id [int]` - номер ОЭС или региона\
`unit_name  [int]` - название учетной единицы (региона, ОЭС)\
`e_use_fact [float]` - объем потребления электроэнргии\
`dt_input [datetime]` - дата и время загрузки информации\
`source [str]` - источник загрузки

## Структура хранения социально-экономических и иных показателей 
Необходимо структурировать информацию о социально-экономических показателях. Формат должен позволять легко добавлять новые показатели. В перспективе возможно использование базы данных.

**df_indicators_db**:\
`index [int]` - номер записи\
`indicator_name [str]` - название индикатора\
`unit_level [int]` - уровень учета - {0:'Россия', 1:'ОЭС', 2:'Регион'}\
`unit_name  [int]` - название учетной единицы (региона, ОЭС)\
`power_sys_id [int]` - номер ОЭС или региона\
`date_time [datetime]` - точка времени в интервале, в котором рассчитан показатель\
`period [str]` - периодичность регистрации показателя ('час', 'месяц', 'год')\
`value [float]` - результат активности в виде набора шкал с ответами\
`moment [datetime]` - дата и время загрузки индикатора\
`source [str]` - источник данных информации о значении индикатора

_Комментарий: `power_sys_id [int]` рассчитывается отдельно в привязке к ОЭС. Важно, что могут быть загружены данные даже по одному показателю из разных источников за разные периоды._

## Алгоритм решения

1. Загружаем исходные данные о потреблении электроэнергии, извлекаем и структурируем  информацию в приведенном выше формате **df_energy_db**
2. Загружаем исходные данные со значениями социально-экономических и других показателей из перечня доступных источников, извлекаем и структурируем их в приведенном выше формате **df_indicators_db**
3. Проверим корреляции показателей с данными о потреблении электроэнергии
4. В рамках модели прогнозирования сделаем среднесрочный прогноз. Сравним данные с информацией Минэкономразвития. Выделим обучающую и тестовую выборки, обучим модель и оценим качество прогноза.

##  СО ЕЭС | Температура в ЕЭС России

https://so-ups.ru/index.php?id=ees_temperature

Данные о средней температуре наружного воздуха в Единой энергетической системе (ЕЭС) России и объединенных энергетических системах (ОЭС) на каждые сутки

```python
# Скачиваем информацию о температуре с сайта https://so-ups.ru/index.php?id=ees_temperature в каталог `temperature`

date_get = dt.datetime(2011, 1, 31)
now = dt.datetime.utcnow() + dt.timedelta(hours=+3)

while date_get < (now + dt.timedelta(days=1)):
    for ees_oes in oes_dict:
        csv_file = (f"https://so-ups.ru/index.php?id=ees_temperature&no_cache=1&tx_ms1cdu_pi1[kpo]={str(ees_oes)}&" + 
                    f"tx_ms1cdu_pi1[dt]={date_get.strftime('%d.%m.%Y')}&tx_ms1cdu_pi1[format]=csv")
        if not f"{date_get.strftime('%Y-%m-%d')}_0_0_{str(ees_oes)}.csv" in os.listdir("temperatures"):
            wget.download(csv_file,'temperatures', None)
    if (date_get < now - dt.timedelta(days=28)) or date_get==now:
        date_get += dt.timedelta(days=28)
    else:
        date_get = now
print(f"Всего файлов со значениями температуры: {len(os.listdir('temperatures'))}")
```

```python
# Загружаем значения температуры на каждый день с 2011 года по всем ОЭС и по ЕЭС

df_temperatures = pd.DataFrame()
for csv_file in os.listdir("temperatures"):
    if csv_file[-3:]=='csv':
        df_temperatures = df_temperatures.append(
            pd.read_csv(f"temperatures\\{csv_file}", sep=';', encoding='cp1251')
            .assign(power_sys_id=int(csv_file[:-4].split('_')[-1])),
            ignore_index=True).drop_duplicates()
df_temperatures['date']=pd.to_datetime(df_temperatures.iloc[:,0], format='%d-%m-%Y')
df_temperatures['t']=df_temperatures.iloc[:,1]
df_temperatures=df_temperatures[['date','power_sys_id','t']]
```
